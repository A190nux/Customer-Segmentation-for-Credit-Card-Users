#Data Preprocessing:

we fill the null values with median as mean would not be accurate nor representative of the data.

```python
# drop this specific row
df_copy.drop(df_copy[df_copy['CREDIT_LIMIT'].isnull()].index, inplace=True)
```

```python
# Create a new dataframe with only the columns we want to consider for the minimum
columns_to_consider = ['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 
                       'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PAYMENTS']

# For each row with missing MINIMUM_PAYMENTS, replace with the minimum of the specified columns
# First, create a mask for rows with missing MINIMUM_PAYMENTS
missing_min_payments_mask = df_copy['MINIMUM_PAYMENTS'].isna()

# For those rows, calculate the minimum of the specified columns
min_values = df_copy.loc[missing_min_payments_mask, columns_to_consider].min(axis=1)

# Replace the missing values with the calculated minimums
df_copy.loc[missing_min_payments_mask, 'MINIMUM_PAYMENTS'] = min_values
```

```python
# List of columns to transform
columns_to_transform = ['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 
                       'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PAYMENTS', 'CREDIT_LIMIT', 'MINIMUM_PAYMENTS']

# Apply log transformation (adding 1 to handle zeros)
for column in columns_to_transform:
    df_copy[column] = np.log1p(df_copy[column])
```

```python
plt.figure(figsize=(16, 12))
plt.subplots_adjust(top=0.9, wspace=0.4, hspace=0.8)

for i in range(len(df_copy.columns)):
    plt.subplot(4, 5, i + 1)
    sns.boxplot(y=df_copy.iloc[:, i], color='red')
    plt.title(df_copy.columns[i])
```

```python
df_scaled = df_copy.copy()
```

# Model training:

```python
from sklearn.cluster import KMeans

square_sum = []
K = range(1, 11)

kmeans_set = {
    'init': 'k-means++',
    'n_init': 8,
    'max_iter': 300,
    'random_state': 42
}
df_scaled.columns = df_scaled.columns.astype(str)
for k in K:
    kmeans = KMeans(n_clusters=k, **kmeans_set)
    kmeans.fit(df_scaled)
    square_sum.append(kmeans.inertia_)

# Plot the elbow
plt.figure(figsize=(8, 5))
plt.plot(K, square_sum, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(K)
plt.grid(True)
plt.show()
```

```python
square_sum
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
Principal_2 = pca.fit_transform(df_scaled)
pca2 = pd.DataFrame(data = Principal_2, columns = ['pca1', 'pca2'])
```

```python
k = 3
kmeans = KMeans(n_clusters=k, **kmeans_set)
kmeans.fit(df_scaled)
pca2['Cluster'] = kmeans.labels_
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=pca2,
    x='pca1',
    y='pca2',
    hue='Cluster',
    palette='Set1',
    s=60,
    alpha=0.7
)
plt.title(f'KMeans Clusters (k={k}) on PCA-Reduced Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
df_scaled['cluster'] = kmeans.labels_
cluster_profiles = df_scaled.groupby('cluster').mean()
print(cluster_profiles)
```
